GAMs allow us to fit a non-linear function to each independent variable, so that we can automatically model non-linear relationships that standard linear regression will miss.

```r
train.data <- read.csv(file.path("..", "data", "training_data.csv"))
test.data <- read.csv(file.path("..", "data", "test_data.csv"))
```

```r
#set seed for reproduction
set.seed(100)

#searching for input variables with zero variance (without the varible Intensity)
x <- train.data[, -c(1,2,3)]
idx.zero.var <- apply(x, 2, var) == 0
x <- x[,!idx.zero.var]

#attache Intensity as factor
x$Intensity <- as.numeric(train.data$Intensity) -1 
y <- train.data$VALENCE.PLEASANTNESS

```

```r
ForwardPredictors = x %>% select(MATS7i,GATS5m,Eig08_EA.ri.,Mor13i,Mor27i, nRCOOR,C.026, H.052,F01.C.S.)
ForwardPredictors$VALENCE.PLEASANTNESS = y
```

```r
creat.formula <- function(x.f, df, y.f = 'VALENCE.PLEASANTNESS'){
    len <- length(df)
    y.f = y.f

    x.f <- sapply(1:len, function(i){
        paste("ns(", x.f[i], ",", "df = ", df[i], ")")
    })
    as.formula(paste(y.f, paste(x.f, collapse = " + "), sep = " ~ "))
}
```

```r
library(boot)
```

```r
set.seed(123)
d <- c(1,2,3)
df <- expand.grid(d,d,d,d,d,d,d,d,d) 
n.deg <- length(df[,1])
cvs <- rep(NA, n.deg)

for (i in 1:n.deg) { 
        formula <- creat.formula(names(ForwardPredictors%>% select(-VALENCE.PLEASANTNESS)) , df[i,])
        #regression
        fit <- glm(formula, data = ForwardPredictors)
        cvs[i] <- cv.glm(ForwardPredictors, fit, K = 10)$delta[1]
}


```
